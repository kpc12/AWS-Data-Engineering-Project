```python
str1 = 'ab#cd!ef'
output = ''
alphas = [char for char in reversed(str1) if char.isalpha()]
# ['s','i','e','m','a','n','y','m']
i = 0 
for char in str1:
    if char.isalpha():
        output += alphas[i]
        i+=1
    else:
        output+=char
print(output)
```

```python
# String Slices

s = "abc"
n = len(s)
substrings = []

for i in range(n):
    for j in range(i+1,n+1):
        substrings.append(s[i:j])
print(substrings)

rev = s[::-1]
# print(rev)
n = len(rev)
substrings = []

for i in range(n):
    for j in range(i+1,n+1):
        substrings.append(rev[i:j])
print(substrings)
```
