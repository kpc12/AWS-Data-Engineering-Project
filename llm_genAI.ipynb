```python
# creating prompt
from langchain_core.prompts import PromptTemplate
prompt_template_name = PromptTemplate(
    input_variables = ['cuisine'],
    template = "I want to open a restaurant for {cuisine} food. Suggest a fancy name for this"
)
prompt_template_name.format(cuisine="Mexican")

```

```python
from langchain_core.prompts import PromptTemplate
chain = prompt_template_name | llm
result = chain.invoke({"cuisine": "American"})
print(result)
```

```python
from langchain_core.prompts import PromptTemplate
food_chain_items = PromptTemplate(
    input_variables = ['restaurant'],
    template = "Suggest some menu items for {restaurant}. Return it as a comma separated list"
)
food_chain_items.format(restaurant="American")
```

```python
#prompt chaining
from langchain_core.prompts import PromptTemplate
from langchain_openai import OpenAI

llm = OpenAI()

# Step 1
prompt1 = PromptTemplate.from_template(
    "Suggest a name for a {cuisine} restaurant"
)

# Step 2
prompt2 = PromptTemplate.from_template(
    "Suggest some menu items for {restaurant}. Return it as a comma separated list"
)

# Chain them
chain = prompt1 | llm | prompt2 | llm

# Run
result = chain.invoke({"cuisine": "Indian"})
print(result)
```
